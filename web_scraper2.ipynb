Scraping TikTok Videos

```python
resultsTab = driver.find_element(By.ID, 'tabs-0-panel-search_video')

posts = resultsTab.find_elements(By.CLASS_NAME, 'e19c29qe10')
posts_size = len(posts)

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    posts = resultsTab.find_elements(By.CLASS_NAME, 'e19c29qe10')
    if len(posts) == posts_size:
        break
    else:
        posts_size = len(posts)
```

```python
for post in posts:
    # Getting video ID
    tiktokIDclass = post.find_element(By.CLASS_NAME, 'e1cg0wnj1')
    tiktokID = tiktokIDclass.find_element(By.TAG_NAME, 'a').get_attribute('href')
    
    if tiktokID in seen:
        continue
    seen.add(tiktokID)

    # Getting video hashtags
    postTags = []
    tagElements = post.find_elements(By.CSS_SELECTOR, 'a[href^="/tag/"]')
    for tagElement in tagElements:
        tag = tagElement.get_attribute('href')
        postTags.append(tag.replace('https://www.tiktok.com/tag/', ''))

    # Removing duplicates
    postTags = list(dict.fromkeys(postTags))

    # Removing single tags
    if len(postTags) <= 1:
        continue
    
    # Transforming all characters to lower case
    postTags = [tag.lower() for tag in postTags]
    
    # If #fitness not in tags, skip
    # if '#gaming' not in postTags:
    #     continue

    linkedTags.append(postTags)
```
